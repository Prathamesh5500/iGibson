```python
from gibson2.envs.behavior_env import BehaviorEnv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
env = BehaviorEnv(config_file='../configs/behavior/behavior_re-shelving_library_books.yaml',
                      mode='headless',
                      action_timestep=1.0 / 10.0,
                      physics_timestep=1.0 / 240.0,
                      )

env.reset()
```

```python
env.simulator.renderer.set_camera([0,-1.9,1.5], [0,-2.9,1.2], [0,0,1], cache=True)
env.simulator.renderer.set_camera([0,-2,1.5], [0,-3,1.2], [0,0,1], cache=True)
rgb, normal, depth, sem_seg, ins_seg, optical_flow = env.simulator.renderer.render(modes=('rgb', 'normal', '3d', 'seg', 'ins_seg', 'optical_flow'))
```

```python
import flow_vis
flow_color = flow_vis.flow_to_color(optical_flow[:,:,:2], convert_to_bgr=False)
plt.imshow(flow_color)
```

```python
sem_image_colored = np.zeros((512,512,4))
for i, item in enumerate(np.unique((sem_seg[:,:,0] * 255).astype(np.uint8))):
    #if item != 0:
    sem_image_colored[(sem_seg[:,:,0] * 255).astype(np.uint8) == item, :] = 0.7 * np.array(get_cmap(len(np.unique((sem_seg[:,:,0] * 255).astype(np.uint8))))(i)) + 0.3
```

```python
Image.fromarray((255 * sem_image_colored).astype(np.uint8)[:,:,:3]).save('behavior_sem_seg.jpg')
```

```python
ins_image_colored = np.zeros((512,512,4))
for i, item in enumerate(np.unique((ins_seg[:,:,0] * 255).astype(np.uint8))):
    ins_image_colored[(ins_seg[:,:,0] * 255).astype(np.uint8) == item, :] = 0.7 * np.array(get_cmap(len(np.unique((ins_seg[:,:,0] * 255).astype(np.uint8))))(i)) + 0.3
```

```python
Image.fromarray((255 * ins_image_colored).astype(np.uint8)[:,:,:3]).save('behavior_ins_seg.jpg')
```

```python
normalized_depth = (depth[:,:,2] - np.min(depth[(depth[:,:,2] != 0), 2])) / (np.max(depth[(depth[:,:,2] != 0), 2]) - np.min(depth[(depth[:,:,2] != 0), 2]))
```

```python
colored_depth = normalized_depth[:,:,None] * np.array([0,0,0,1]) + (1-normalized_depth[:,:,None]) * np.array([1,1,1,1])
colored_depth[(depth[:,:,2] == 0), :] = [0,0,0,1]
```

```python
Image.fromarray((255 * colored_depth).astype(np.uint8)[:,:,:3]).save('behavior_depth.jpg')
```
