```python
def post_processing_eye_tracking_data(data):
    """Post processing eye tracking data to extract fixations"""
    threshold = 0.0
    for n in range(data.shape[0]):
        for i in range(1, data.shape[1]):
            if np.abs(data[n][i] - data[n][i-1]) < threshold:
                data[n][i] = data[n][i - 1]
    return data
```
